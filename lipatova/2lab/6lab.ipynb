# ML: Р—Р°РґР°С‡Р° Р°СЃСЃРѕС†РёР°С†РёРё : РњРµС‚РѕРґ Apriori

Р”Р»СЏ СЂРµС€РµРЅРёСЏ Р·Р°РґР°С‡Рё РјР°С€РёРЅРЅРѕРіРѕ РѕР±СѓС‡РµРЅРёСЏ С‚СЂРµР±СѓРµС‚СЃСЏ:

1) РґР°РЅРЅС‹Рµ - РґР°С‚Р°СЃРµС‚

2) РѕРїСЂРµРґРµР»РёС‚СЊ С‚РёРї СЂРµС€Р°РµРјРѕР№ Р·Р°РґР°С‡Рё - Р°СЃСЃРѕС†РёР°С†РёСЏ

3) РѕРїСЂРµРґРµР»РёС‚СЊ РјРµСЂС‚РёРєРё РѕС†РµРЅРєРё - support, confidence, lift, leverage, conviction
Р§Р°СЃС‚РѕС‚Р°: Support (С‡Р°СЃС‚РѕС‚Р° РєРѕРјР±РёРЅР°С†РёР№).
РЎРёР»Р° СЃРІСЏР·Рё: Confidence (РєР°Рє С‡Р°СЃС‚Рѕ РїСЂР°РІРёР»Рѕ РІС‹РїРѕР»РЅСЏРµС‚СЃСЏ).
Р—РЅР°С‡РёРјРѕСЃС‚СЊ СЃРІСЏР·Рё: Lift (РЅР°СЃРєРѕР»СЊРєРѕ СЃРёР»СЊРЅРѕ РїРµСЂРµРјРµРЅРЅС‹Рµ СЃРІСЏР·Р°РЅС‹).
Leverage РЅР°СЃРѕРєР»СЊРєРѕ РёРЅС‚РµСЂРµСЃРЅРѕР№ РјРѕР¶РµС‚ Р±С‹С‚СЊ Р±РѕР»РµРµ РІС‹СЃРѕРєР°СЏ С‡Р°СЃС‚РѕС‚Р° AT

5) РѕРїСЂРµРґРµР»РёС‚СЊ СЃРїРѕСЃРѕР± РІРёР·СѓР°Р»РёР·Р°С†РёРё СЂРµР·СѓР»СЊС‚Р°С‚РѕРІ - matplotlib

6) РѕРїСЂРµРґРµР»РёС‚СЊ РјРµС‚РѕРґ СЂРµС€РµРЅРёСЏ (РѕРїС‚РёРјРёР·Р°С†РёРё) - Apriori

# Р‘РёР±Р»РёРѕС‚РµРєР° Mlxtend (machine learning extensions)

```python
# РџРѕРґРєР»СЋС‡Р°РµРј Р±РёР±Р»РёРѕС‚РµРєРё
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Р—Р°РіСЂСѓР·РєР° РґР°РЅРЅС‹С…
df = pd.read_csv(r"C:\Users\tmas2\breast-cancer.csv")

# Р’С‹Р±РёСЂР°РµРј С‡РёСЃР»РѕРІС‹Рµ РєРѕР»РѕРЅРєРё РґР»СЏ Р°РЅР°Р»РёР·Р°, РѕРіСЂР°РЅРёС‡РёРІ РёС… РєРѕР»РёС‡РµСЃС‚РІРѕ
numerical_columns = df.columns[2:7]  # РџСЂРѕРїСѓСЃРєР°РµРј 'id' Рё 'diagnosis', Р±РµСЂРµРј 5 РєРѕР»РѕРЅРѕРє РґР»СЏ Р°РЅР°Р»РёР·Р°

# Р’С‹С‡РёСЃР»СЏРµРј РјРµРґРёР°РЅС‹ РґР»СЏ РІСЃРµС… РІС‹Р±СЂР°РЅРЅС‹С… С‡РёСЃР»РѕРІС‹С… РєРѕР»РѕРЅРѕРє
medians = df[numerical_columns].median()

# РЎРѕР·РґР°РµРј СЃРїРёСЃРѕРє С‚СЂР°РЅР·Р°РєС†РёР№
transactions = []

# РџСЂРµРѕР±СЂР°Р·СѓРµРј РґР°РЅРЅС‹Рµ РІ С‚СЂР°РЅР·Р°РєС†РёРё
for index, row in df.iterrows():
    transaction = []
    for col in numerical_columns:
        # Р Р°Р·РґРµР»СЏРµРј РЅР° РєР°С‚РµРіРѕСЂРёРё "high" Рё "low" РѕС‚РЅРѕСЃРёС‚РµР»СЊРЅРѕ РјРµРґРёР°РЅС‹ СЃС‚РѕР»Р±С†Р°
        if row[col] > medians[col]:
            transaction.append(f"{col}_high")
        else:
            transaction.append(f"{col}_low")
    if transaction:  # Р”РѕР±Р°РІР»СЏРµРј С‚СЂР°РЅР·Р°РєС†РёСЋ РІ СЃРїРёСЃРѕРє, РµСЃР»Рё РѕРЅР° РЅРµ РїСѓСЃС‚Р°СЏ
        transactions.append(transaction)

# РџСЂРµРѕР±СЂР°Р·СѓРµРј СЃРїРёСЃРѕРє С‚СЂР°РЅР·Р°РєС†РёР№ РІ DataFrame
encoder = TransactionEncoder()
encoded_array = encoder.fit(transactions).transform(transactions)
encoded_df = pd.DataFrame(encoded_array, columns=encoder.columns_)

# РџСЂРёРјРµРЅСЏРµРј Р°Р»РіРѕСЂРёС‚Рј Apriori СЃ РјРёРЅРёРјР°Р»СЊРЅРѕР№ РїРѕРґРґРµСЂР¶РєРѕР№ (0.1)
frequent_itemsets = apriori(encoded_df, min_support=0.1, use_colnames=True, low_memory=True)

# Р Р°СЃСЃС‡РёС‚С‹РІР°РµРј РєРѕР»РёС‡РµСЃС‚РІРѕ РЅР°Р±РѕСЂРѕРІ
num_itemsets = len(frequent_itemsets)

# Р“РµРЅРµСЂР°С†РёСЏ РїСЂР°РІРёР» Р°СЃСЃРѕС†РёР°С†РёРё СЃ РёСЃРїРѕР»СЊР·РѕРІР°РЅРёРµРј РІР°С€РµР№ РІРµСЂСЃРёРё association_rules
rules = association_rules(
    df=frequent_itemsets,
    num_itemsets=num_itemsets,
    metric="confidence",
    min_threshold=0.6
)

# Р’С‹РІРѕРґРёРј СЂРµР·СѓР»СЊС‚Р°С‚С‹
print("Р§Р°СЃС‚Рѕ РІСЃС‚СЂРµС‡Р°СЋС‰РёРµСЃСЏ РЅР°Р±РѕСЂС‹ СЌР»РµРјРµРЅС‚РѕРІ:")
print(frequent_itemsets)

print("\nРљРѕР»РёС‡РµСЃС‚РІРѕ РЅР°Р№РґРµРЅРЅС‹С… РЅР°Р±РѕСЂРѕРІ:", num_itemsets)

print("\nРђСЃСЃРѕС†РёР°С‚РёРІРЅС‹Рµ РїСЂР°РІРёР»Р°:")
print(rules)
```

# ML: Р—Р°РґР°С‡Р° Р°СЃСЃРѕС†РёР°С†РёРё : РњРµС‚РѕРґ FP-Growth

## РђР»РіРѕСЂРёС‚Рј fpgrowth

> *Frequent Pattern-Growth (FPG) - РІС‹СЂР°С‰РёРІР°РЅРёРµ РїРѕРїСѓР»СЏСЂРЅС‹С… (С‡Р°СЃС‚Рѕ РІСЃС‚СЂРµС‡Р°СЋС‰РёС…СЃСЏ) РїСЂРµРґРјРµС‚РЅС‹С… РЅР°Р±РѕСЂРѕРІ*

FP-Growth РѕС‚Р»РёС‡Р°РµС‚СЃСЏ РѕС‚ Apriori С‚РµРј, С‡С‚Рѕ РЅРµ  С‚СЂРµР±СѓРµС‚ РіРµРЅРµСЂР°С†РёРё РєР°РЅРґРёРґР°С‚РѕРІ. РњРµС‚РѕРґ РёСЃРїРѕР»СЊР·СѓРµС‚ СЃС‚СЂСѓРєС‚СѓСЂСѓ РґР°РЅРЅС‹С… С‚Р°Рє РЅР°Р·С‹РІР°РµРјРѕРіРѕ FP-РґРµСЂРµРІР° (РґРµСЂРµРІРѕ С‡Р°СЃС‚С‹С… РїР°С‚С‚РµСЂРЅРѕРІ) Р±РµР· СЏРІРЅРѕР№ РіРµРЅРµСЂР°С†РёРё РЅР°Р±РѕСЂРѕРІ РєР°РЅРґРёРґР°С‚РѕРІ (РїРѕРґС…РѕРґРёС‚ РґР»СЏ Р±РѕР»СЊС€РёС… РЅР°Р±РѕСЂРѕРІ РґР°РЅРЅС‹С…).


### Р‘РёР±Р»РёРѕС‚РµРєР° Mlxtend

```python
# РџРѕРґРєР»СЋС‡Р°РµРј РЅРµРѕР±С…РѕРґРёРјС‹Рµ Р±РёР±Р»РёРѕС‚РµРєРё
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Р—Р°РіСЂСѓР¶Р°РµРј РґР°РЅРЅС‹Рµ
df = pd.read_csv(r'C:\Users\tmas2\breast-cancer.csv')

# РЈР±РёСЂР°РµРј СЃС‚РѕР»Р±РµС† ID, РµСЃР»Рё РѕРЅ РЅРµ РЅСѓР¶РµРЅ РґР»СЏ Р°РЅР°Р»РёР·Р°
df.drop(columns=['id'], inplace=True)

# РџСЂРµРѕР±СЂР°Р·СѓРµРј РєРѕР»РёС‡РµСЃС‚РІРµРЅРЅС‹Рµ РїСЂРёР·РЅР°РєРё РІ РєР°С‚РµРіРѕСЂРёР°Р»СЊРЅС‹Рµ
# РќР°РїСЂРёРјРµСЂ, РјРѕР¶РЅРѕ СЂР°Р·РґРµР»РёС‚СЊ РёС… РЅР° "РІС‹СЃРѕРєРёРµ" Рё "РЅРёР·РєРёРµ" Р·РЅР°С‡РµРЅРёСЏ РѕС‚РЅРѕСЃРёС‚РµР»СЊРЅРѕ РјРµРґРёР°РЅС‹
df_binned = df.copy()
for column in df.columns:
    if column not in ['diagnosis']:  # РџСЂРѕРїСѓСЃРєР°РµРј РєР°С‚РµРіРѕСЂРёР°Р»СЊРЅС‹Р№ СЃС‚РѕР»Р±РµС†
        df_binned[column] = pd.cut(df[column], bins=2, labels=["Low", "High"])

# Р”РѕР±Р°РІР»СЏРµРј РґРёР°РіРЅРѕР· РІ РєР°С‡РµСЃС‚РІРµ РѕС‚РґРµР»СЊРЅРѕР№ С‚СЂР°РЅР·Р°РєС†РёРё
df_binned['diagnosis'] = df['diagnosis']

# РџСЂРµРѕР±СЂР°Р·СѓРµРј РґР°РЅРЅС‹Рµ РІ С„РѕСЂРјР°С‚ С‚СЂР°РЅР·Р°РєС†РёР№
transactions = df_binned.apply(lambda x: x.tolist(), axis=1).tolist()

# РљРѕРґРёСЂСѓРµРј РґР°РЅРЅС‹Рµ С‡РµСЂРµР· TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)

# РџСЂРёРјРµРЅСЏРµРј FP-Growth РґР»СЏ РЅР°С…РѕР¶РґРµРЅРёСЏ С‡Р°СЃС‚РѕС‚РЅС‹С… РЅР°Р±РѕСЂРѕРІ СЌР»РµРјРµРЅС‚РѕРІ
frequent_itemsets = fpgrowth(df_transactions, min_support=0.05, use_colnames=True)

# Р”РѕР±Р°РІР»СЏРµРј РєРѕР»РёС‡РµСЃС‚РІРѕ РЅР°Р±РѕСЂРѕРІ СЌР»РµРјРµРЅС‚РѕРІ РІ Р°СЂРіСѓРјРµРЅС‚ РґР»СЏ РїСЂР°РІРёР» Р°СЃСЃРѕС†РёР°С†РёРё
num_itemsets = len(frequent_itemsets)

# Р’С‹РІРѕРґРёРј СЂРµР·СѓР»СЊС‚Р°С‚
print("Р§Р°СЃС‚РѕС‚РЅС‹Рµ РЅР°Р±РѕСЂС‹ СЌР»РµРјРµРЅС‚РѕРІ:")
print(frequent_itemsets)

# Р“РµРЅРµСЂРёСЂСѓРµРј РїСЂР°РІРёР»Р° Р°СЃСЃРѕС†РёР°С†РёРё
rules = association_rules(
    df=frequent_itemsets,
    num_itemsets=num_itemsets,
    metric="confidence",
    min_threshold=0.8
)

# Р’С‹РІРѕРґРёРј РїСЂР°РІРёР»Р° Р°СЃСЃРѕС†РёР°С†РёРё
print("\nРџСЂР°РІРёР»Р° Р°СЃСЃРѕС†РёР°С†РёРё:")
print(rules)
```

```python
# РџРѕРґРєР»СЋС‡Р°РµРј РЅРµРѕР±С…РѕРґРёРјС‹Рµ Р±РёР±Р»РёРѕС‚РµРєРё
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Р—Р°РіСЂСѓР¶Р°РµРј РґР°РЅРЅС‹Рµ
df = pd.read_csv(r'C:\Users\tmas2\breast-cancer.csv')

# РЈР±РёСЂР°РµРј СЃС‚РѕР»Р±РµС† ID, С‚Р°Рє РєР°Рє РѕРЅ РЅРµ РЅРµСЃРµС‚ РїРѕР»РµР·РЅРѕР№ РёРЅС„РѕСЂРјР°С†РёРё РґР»СЏ Р°РЅР°Р»РёР·Р°
df.drop(columns=['id'], inplace=True)

# РџСЂРµРѕР±СЂР°Р·СѓРµРј С‡РёСЃР»РѕРІС‹Рµ РїСЂРёР·РЅР°РєРё РІ РєР°С‚РµРіРѕСЂРёР°Р»СЊРЅС‹Рµ (РЅР°РїСЂРёРјРµСЂ, Low Рё High)
df_binned = df.copy()
for column in df.columns:
    if column not in ['diagnosis']:  # РџСЂРѕРїСѓСЃРєР°РµРј РєР°С‚РµРіРѕСЂРёР°Р»СЊРЅС‹Р№ СЃС‚РѕР»Р±РµС†
        df_binned[column] = pd.cut(df[column], bins=2, labels=["Low", "High"])

# РџСЂРµРѕР±СЂР°Р·СѓРµРј РєР°Р¶РґСѓСЋ СЃС‚СЂРѕРєСѓ РІ РЅР°Р±РѕСЂ РєР°С‚РµРіРѕСЂРёР№ (С‚СЂР°РЅР·Р°РєС†РёСЋ)
transactions = df_binned.apply(lambda x: x.tolist(), axis=1).tolist()

# РљРѕРґРёСЂСѓРµРј С‚СЂР°РЅР·Р°РєС†РёРё СЃ РїРѕРјРѕС‰СЊСЋ TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)

# РџСЂРёРјРµРЅСЏРµРј FP-Growth РґР»СЏ РЅР°С…РѕР¶РґРµРЅРёСЏ С‡Р°СЃС‚РѕС‚РЅС‹С… РЅР°Р±РѕСЂРѕРІ СЌР»РµРјРµРЅС‚РѕРІ
frequent_itemsets = fpgrowth(df_transactions, min_support=0.05, use_colnames=True)

# Р“РµРЅРµСЂРёСЂСѓРµРј Р°СЃСЃРѕС†РёР°С‚РёРІРЅС‹Рµ РїСЂР°РІРёР»Р°
rules = association_rules(
    df=frequent_itemsets,
    num_itemsets=len(frequent_itemsets),
    metric="confidence",
    min_threshold=0.5
)

# Р’С‹РІРѕРґРёРј СЂРµР·СѓР»СЊС‚Р°С‚С‹
print("Р§Р°СЃС‚Рѕ РІСЃС‚СЂРµС‡Р°СЋС‰РёРµСЃСЏ РЅР°Р±РѕСЂС‹ СЌР»РµРјРµРЅС‚РѕРІ:")
print(frequent_itemsets)

print("\nРђСЃСЃРѕС†РёР°С‚РёРІРЅС‹Рµ РїСЂР°РІРёР»Р°:")
print(rules)
```
